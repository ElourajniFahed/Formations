```
datardd.groupBy(lambda x : (x["HOUR"])).map(lambda x : (x[0],list(x[1]))).map(lambda x : (x[0],len(x[1]))).reduce(lambda x,y : ("Maxe vale",max(x[1],y[1]))) # a chaque foi que je lancer cette operation le rdd vas se baser sur son lineage et donc on vas perdre le temps
```

```
rdd_persisted =datardd.groupBy(lambda x : (x["HOUR"])).map(lambda x : (x[0],list(x[1]))).map(lambda x : (x[0],len(x[1]))).persist()
# persist allow us to save the rdd in memory of the node that willi execute it 
```

```
rdd_persisted.reduce(lambda x,y : ("Maxe vale",max(x[1],y[1]))) # the first time it will take time to execute ans to be saved on the node Memory
# from the second time of execution it will be much faster
```

```
rdd_cached =datardd.groupBy(lambda x : (x["HOUR"])).map(lambda x : (x[0],list(x[1]))).map(lambda x : (x[0],len(x[1]))).cache()
# cache allow us to save the rdd in memory of the node that willi execute it 
```

```
rdd_cached.reduce(lambda x,y : ("Maxe vale",max(x[1],y[1]))) # persisted rdd is much faster than Cached Rdd
```
